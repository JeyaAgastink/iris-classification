## Relationship Of The Data Features With The Target



```python
data['target_names']
```

```python
#create new column with the species name
df['target_names'] = df['target'].map({0: "setona", 1: "versicolor", 2: "virginica"})
```

```python
col = "sepal length (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
```

```python
col = "sepal width (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
```

```python
col = "petal length (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
```

```python
col = "petal width (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
```
